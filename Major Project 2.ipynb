```python
#4.divide the into input
x = df.iloc[:,3:5].values
x
```

```python
#VISULAISTION 
import matplotlib.pyplot as plt
plt.scatter(df['prep_time'],df['cook_time'])
#Here we have got only one clustering before applying any clustering technique
```

```python
#Here our main task is to find out the number of clusters(k)
import numpy as np
np.sqrt(255) # 255 is the total no of points
#No of cluster - k
```

```python
#1.Elbow method
from sklearn.cluster import KMeans
k = range(2,16)

sse = []#blank list

#for i in range(2,16):
for i in k:
  model_demo = KMeans(n_clusters = i,random_state = 0)
  model_demo.fit(x)
  sse.append(model_demo.inertia_)
plt.scatter(k,sse)
plt.plot(k,sse)


```

```python
#we will now considere the point at which the elbow is more prominent
```

```python
#2.SILHOUETTE SCORE METHOD
from sklearn.metrics import silhouette_score
k = range(2,16)
for i in k:
  model_demo = KMeans(n_clusters = i,random_state = 0)
  model_demo.fit(x)
  y_pred = model_demo.predict(x)
  print(f"{i} Clusters ,Score = {silhouette_score(x,y_pred)}")
  plt.bar(i,silhouette_score(x,y_pred))
```

```python
#CONFIRMATION : THE NO OF CLUSTERS TO BE CONSIDERED IS 3
```

```python
#7.APPLY CLUSTER
k = 3
from sklearn.cluster import KMeans

model = KMeans(n_clusters = k,random_state = 0)
model.fit(x)

KMeans(n_clusters=3, random_state=0)
```
